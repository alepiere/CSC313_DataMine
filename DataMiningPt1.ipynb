```python
#Q2
#The data-set contains many Assignments, and each Assignment is made up of multiple Problem. Which Problem had the highest average number of attempts per student?
# m_table['ProblemID']
early_df = pd.read_csv("S19_All_Release_2_10_22/early.csv")
late_df = pd.read_csv("S19_All_Release_2_10_22/late.csv")

# Concatenate the two DataFrames
df = pd.concat([early_df, late_df])

# Find the problem with the highest average number of attempts per student
problem_with_highest_average_attempts = df.groupby('ProblemID')["Attempts"].mean().idxmax()
highest_average_attempts = df.groupby('ProblemID')["Attempts"].mean().max()

print("Problem with the highest average number of attempts per student:", problem_with_highest_average_attempts)
print("Average number of attempts per student for this problem:", highest_average_attempts)

#102
```

```python
#Q3
#On which Problem did students tend to face the most compiler errors?

#compute, for each problem, the average number of compiler errors across all students, and then report the maximum, or
#compute the total number of compiler errors faced on each problem, and report the maximum

df_main = pd.read_csv("S19_All_Release_2_10_22/Data/MainTable.csv")


# Find the problem with the highest average number of errors per student

df_main[["SubjectID", "ProblemID", "Compile.Result"]]

failures = df_main[df_main['Compile.Result'] == 'Error']
error_count = failures.groupby(['SubjectID', 'ProblemID']).size().reset_index(name='ErrorCount')
average_errors = error_count.groupby('ProblemID')['ErrorCount'].mean().reset_index()
problem_with_max_errors = average_errors.loc[average_errors['ErrorCount'].idxmax()]

problem_with_max_errors

# Which one would be more meaningful?

# The average per student is more meaningful. We need to account for the number of students who have attempted the
# problem, because otherwise a problem with hardly any attempts but many errors per attempt will not be considered
# by calculating only totals.
```
